```python
class Song:
    key_signature = None;
    time_signature = None;
    
```

```python
from mido import MidiFile
mid = MidiFile('Super Mario 64 - Medley.mid')
supermario = Song();

for i, track in enumerate(mid.tracks):
    print('Track {}: {}'.format(i, track.name))
    for msg in track:
        print(msg)
        if msg.type == "time_signature":
            supermario.time_signature = msg
        elif msg.type == "key_signature":
            supermario.key_signature = msg;
    
```
